```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_state_and_union_territory_capitals_in_India'
html = requests.get(url)
htmlContent = html.content

bSoup = BeautifulSoup(htmlContent, 'html.parser')
```

```python
type(bSoup)
```

```python
table = bSoup.find('table',{'class':'wikitable sortable plainrowheaders'})
```

```python
type(table)
```

```python
SNoList =[]
StateList =[]
AdminCapitaList =[]
LegisCapitalList =[]
JudisCapitalList =[]
YearList =[]
FormerCapitalList =[]

for r in table.findAll('tr'):
    col = r.findAll('th')
    data = r.findAll('td')
    if len(col) == 7:
        Sno = col[0].text.replace('\n',' ')
        State = col[1].text.replace('\n',' ')
        AdminCapital = col[2].text.replace('\n',' ')
        LegisCapital = col[3].text.replace('\n',' ')
        JudisCapital = col[4].text.replace('\n',' ')
        Year = col[5].text.replace('\n',' ')
        FormerCapital = col[6].text.replace('\n',' ')
   
    if len(data) == 5:
        Sno = col[0].text.replace('\n',' ')
        State = col[1].text.replace('\n',' ')
        AdminCapital = data[0].text.replace('\n',' ')
        LegisCapital = data[1].text.replace('\n',' ')
        JudisCapital = data[2].text.replace('\n',' ')
        Year = data[3].text.replace('\n',' ')
        FormerCapital = data[4].text.replace('\n',' ')

    SNoList.append(Sno)
    StateList.append(State)
    AdminCapitaList.append(AdminCapital)
    LegisCapitalList.append(LegisCapital)
    JudisCapitalList.append(JudisCapital)
    YearList.append(Year)
    FormerCapitalList.append(FormerCapital)
```

```python
myList = np.column_stack([SNoList,StateList, AdminCapitaList, LegisCapitalList, JudisCapitalList, YearList, FormerCapitalList])

resultTable = pd.DataFrame(myList)
```

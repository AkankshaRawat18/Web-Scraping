## WebScraping Critic Ratings from Rotten Tomatoes

### Loading Packages

```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import re
import string
```

### Scraping Critic Name, Movie Name, Critic Rating and Year

```python
alphaList = list(string.ascii_lowercase)
criticRating = 100 # Default Value
movieNameList = []
criticRatingList = []
yearList = []
criticNameList = []

for alpha in range(0, len(alphaList)):
    url = 'https://www.rottentomatoes.com/critics/authors?letter='+alphaList[alpha]
    
    html = requests.get(url)
    htmlContent = html.content
    bSoupRT = BeautifulSoup(htmlContent,'html.parser')
    
    criticTable = bSoupRT.find('table',{'class':'table table-striped borderless'})
    
    criticUrlList = []
    for row in criticTable.findAll('p',{'class':'critic-names'}):
        criticName = row.find('a').text.lower().replace(' ','-').replace('.','')
        criticUrlList.append(criticName)
        
    baseUrl = 'https://www.rottentomatoes.com/critic/'


    for i in range(0,len(criticUrlList)):
        url = baseUrl+criticUrlList[i]+'/movies'
        html = requests.get(url)
        htmlContent = html.content

        bSoupRT = BeautifulSoup(htmlContent,'html.parser')

        ratingTable = bSoupRT.find('table',{'class':'table table-striped'})


        print(' ------------------',criticUrlList[i])

        criticNameTable = bSoupRT.find('div',{'id':'criticPanel'})
        if criticNameTable != None:
            criticName = criticNameTable.find('h2',{'class':'title'}).text.replace('\n','')

            for row in ratingTable.findAll('tr'):
                name = row.find('td',{'class':'center'})
                if name!= None:
                    movieName = name.find('a',{'class':'movie-link'}).text

                    if movieName=='':
                        result = 'NA'
                    else:
                        year = name.text
                        regexYear = re.compile('\([0-9][0-9][0-9][0-9]\)')
                        result = regexYear.findall(year)

                        if result != None:
                            if len(result)>0:
                                yearVal= result[0].replace('(','').replace(')','')
                            else:
                                yearVal = 'NA'
                        else:
                            yearVal = 'NA'


                    yearList.append(yearVal)
                    movieNameList.append(movieName)

                    criticNameList.append(criticName)

                ratingRow = row.find('td')

                if ratingRow!= None:

                    ratingPos = ratingRow.find('span',{'class':'icon tiny fresh'})
                    if ratingPos!= None:
                        criticRating = 1

                    ratingNeg = ratingRow.find('span',{'class':'icon tiny rotten'})
                    if ratingNeg!= None:
                        criticRating = 0


                    criticRatingList.append(criticRating)
```

```python
movieRating = np.column_stack([criticNameList,movieNameList,criticRatingList,yearList])
movieRatingDF = pd.DataFrame(movieRating)
```

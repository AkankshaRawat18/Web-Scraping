## WebScraping Yield Curves

```python
from urllib.request import urlopen  
from bs4 import BeautifulSoup
import pandas as pd
```

```python
html = urlopen('https://www.treasury.gov/resource-center/data-chart-center/interest-rates/Pages/TextView.aspx?data=yieldYear&year=2018')
```

```python
bsyc = BeautifulSoup(html.read(), "lxml")
```

```python
# This will create a new text file in the current working directory
fout = open('bsyc_temp.txt', 'wt',encoding='utf-8')
```

```python
fout.write(str(bsyc))
```

```python
# The text file will be available for viewing only after this statement
fout.close()
```

```python
print(str(bsyc.chart))
```

```python
print(str(bsyc.table))
```

```python
# so get a list of all table tags
table_list = bsyc.findAll('table')
# table_list
```

```python
# how many are there?
print('there are', len(table_list), 'table tags')
```

```python
# look at the first 50 chars of each table
for t in table_list:
    print(str(t)[:50])
```

```python
# only one class="t-chart" table, so add that
# to findAll as a dictionary attribute
tc_table_list = bsyc.findAll('table', { "class" : "t-chart" } )
```

```python
print(len(tc_table_list), 't-chart tables')
```

```python
tc_table = tc_table_list[0]
```

```python
# what are this table's components/children?
for c in tc_table.children:
    print(str(c)[:50])
```

```python
# tag tr means table row, containing table data
# what are the children of those rows?
for c in tc_table.children:
    for r in c.children:
        print(str(r)[:50])
```

```python
# we have found the table data!
# just get the contents of each cell
for c in tc_table.children:
    for r in c.children:
        print(r.contents)
```

```python

dataFrame = pd.DataFrame()
dataList = []
for c in tc_table.children:
    for r in c.children:
        temp = pd.DataFrame(r.contents)
        dataFrame = pd.concat([dataFrame,temp])

    
```
